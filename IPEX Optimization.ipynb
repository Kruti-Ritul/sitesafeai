```python
import torch
from ultralytics import YOLO
import intel_extension_for_pytorch as ipex

# Load the YOLO model
model = YOLO("best.pt")

# Ensure the model is properly loaded and set to eval mode
model.model.eval()

# Validate the model first (before applying IPEX optimization)
try:
    results = model.val(data="/home/uc17f0f706c08b0f376b3a79fc5f794e/sitesafeai/data.yaml")
    #print("Validation results before optimization:")
    #print(results)
except Exception as e:
    print(f"Error during validation: {e}")

# Now apply IPEX optimization (without fusion)
try:
    model.model = ipex.optimize(model.model, dtype=torch.float32)
    print("Model optimized with IPEX successfully!")
except Exception as e:
    print(f"Error during IPEX optimization: {e}")

# Validate the model again after applying optimization
try:
    results = model.val(data="/home/uc17f0f706c08b0f376b3a79fc5f794e/sitesafeai/data.yaml")
    #print("Validation results after optimization:")
    #print(results)
except Exception as e:
    print(f"Error during validation after optimization: {e}")

# Print the mean results which include precision, recall, mAP scores
mean_results = results.mean_results()
print("Mean Results:", mean_results)

# If you want specific metrics:
print(f"Precision: {mean_results[0]}")
print(f"Recall: {mean_results[1]}")
print(f"mAP@.5: {mean_results[2]}")
print(f"mAP@.5:.95: {mean_results[3]}")
```

```python
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
import torch

# Load the YOLO model
model = YOLO("best.pt")

# Load optimized weights into the model
optimized_weights = torch.load("ipex_optimized_model_weights.pt")
model.model.load_state_dict(optimized_weights)
print("Optimized weights loaded successfully!")

# Path to the image
image_path = "./data/const1.jpg"

# Perform inference
results = model(image_path)

# Annotate and save the image
annotated_image = results[0].plot()  # Annotates the image

# Convert to PIL Image for compatibility
annotated_pil_image = Image.fromarray(annotated_image)

# Display the annotated image
annotated_pil_image.show()  # Opens the image in the default viewer



output_path = "outputs/annotated_image.jpg"
Image.fromarray(annotated_image).save(output_path)  # Save the image
print(f"Annotated image saved to: {output_path}")

# Print summary of results
print("Detection Summary:")
print(results[0].boxes.data)  # Bounding boxes and predictions
```
